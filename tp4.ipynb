```python
from random import random
import numpy as np
from math import exp, sqrt
import time

def udiscrete(m, k):
    """
    Uniform discrete variable in interval [m, k]
    """
    u = random()

    return int(u * (k - m + 1)) + m
```

```python
## Ejercicio 1

N = [100, 1000, 10000, 100000]
R = 10
for n in N:
    suma_exitos = 0
    suma_cuadrados = 0
    r_concidencias = 0
    r_concidencias_ii = 0
    for i in range(n):
        permutacion = np.random.permutation(np.arange(100))
        exito = permutacion == np.arange(100)
        cumple_criterio = True
        for i in range(R):
            if not(exito[i]):
                cumple_criterio = False
                break
        if cumple_criterio:
            r_concidencias += 1

        aciertos = 0    
        for j in range(100):
            if exito[j]:
                aciertos += 1
        if aciertos == R:
            r_concidencias_ii += 1
                
        suma_exitos += np.sum(exito)
        suma_cuadrados += np.sum(exito)^2

    print(n)
    print(f"Esperanza: {suma_exitos / n}")
    print(f"Varianza: {(suma_cuadrados - suma_exitos^2) / n}")
    print(f"Probabilidad de r primeras coincidencias: {r_concidencias/n}")
    print(f"Probabilidad de r exactamente coincidencias: {r_concidencias_ii / n}")
    print('------------------------')
```

```python
## Ejercicio 2
suma=0
Nsim=100
N = 10000
real = 0
estimado = 0
tiempo_estimacion  = time.time()
for _ in range(Nsim):
    U = int(random() * N) + 1
    suma += exp(U/N)
estimado = suma / Nsim * N
tiempo_estimacion  = (time.time() - tiempo_estimacion) * 10**3   # tiempo en ms

tiempo_real  = time.time()
for x in range(N):
    real += exp(x/N)
tiempo_real  = (time.time() - tiempo_real) * 10**3

print(f"Estimado: {estimado} en {tiempo_estimacion} ms")
print(f"Valor Real: {real} en {tiempo_real} ms")
print("Diferencia: {}".format(abs(estimado-real)))
```

```python
##Ejercicio 5
def binomial(n,p):
    """
    Genera un valor de la variable aleatoria binomial usando el metodo de la 
    transformada inversa
    Args:
        N: Cantidad de ensayos independientes
        P: Probabilidad de exito en el ensayo
    """
    c = p / (1 - p)
    prob = (1 - p) ** n
    F = prob; i=0
    U = random()
    while U >= F:
        prob *= c * (n-i) / (i+1)
        F += prob
        i += 1
    return i

def binomial_ensayos(n, p):
    """
    Genera un valor de la variable aleatoria binomial usando el metodo de la 
    realizando N ensayos y contando cuantos de ellos fueron exitosos
    """
    cantidad_exitos = 0
    for _ in range(n):
        U = random()
        if U < p:
            cantidad_exitos += 1
    return cantidad_exitos

def esperanza_binomial(ensayos, p, N, generador):
    """
    Args:
        ensayos   : Parametro de la binomial
        p         : Probabilidad de exito en el ensayo
        N         : Numero de muestras para la estimacion
        generador : Generador de variables aleatorias binomiales
    """
    esperanza = 0
    for _ in range(N):
        esperanza += generador(ensayos, p)
    return esperanza / N

def estimar_p_x_binomial(ensayos, p, N, generador, X):
    p_x = 0
    for _ in range(N):
        B = generador(ensayos, p)
        if B == X:
            p_x += 1
    return p_x / N


### Comparacion de tiempos
n, p = 10, 0.3
N = 10000
# Tiempo de la transformada inversa
tiempo_TI = time.time()
for _ in range(N):
    binomial(n, p)
tiempo_TI = np.round((time.time() - tiempo_TI) * 10**3, 4)

# Tiempo de la binomial simulada
tiempo_SIM = time.time()
for _ in range(N):
    binomial_ensayos(n, p)
tiempo_SIM = np.round((time.time() - tiempo_SIM) * 10**3, 4)

print("""
Tiempo requerido por el metodo de la transformada inversa: {} ms
Tiempo requerido por el metodo de simulacion             : {} ms
""".format(tiempo_TI, tiempo_SIM))



### Comparacion de esperanzas
esperanza_TI  = esperanza_binomial(ensayos = 10, p = 0.3, N = N, generador = binomial)
esperanza_SIM = esperanza_binomial(ensayos = 10, p = 0.3, N = N, generador = binomial_ensayos)

p_b_0_TI   = estimar_p_x_binomial(ensayos = 10, p = 0.3, N = N, generador = binomial, X = 0)
p_b_10_TI  = estimar_p_x_binomial(ensayos = 10, p = 0.3, N = N, generador = binomial, X = 10)
p_b_0_SIM  = estimar_p_x_binomial(ensayos = 10, p = 0.3, N = N, generador = binomial_ensayos, X = 0)
p_b_10_SIM = estimar_p_x_binomial(ensayos = 10, p = 0.3, N = N, generador = binomial_ensayos, X = 10)

print("""
Esperanza obtenida usando transformada inversa: {} 
Esperanza obtenida usando simulacion: {}
""".format(esperanza_TI, esperanza_SIM))

print("""
Probabilidad    Metodo T.I       Metodo SIM
  P(X = 0)      {}           {}
  P(X = 10)     {}              {}
""".format(p_b_0_TI, p_b_0_SIM, p_b_10_TI, p_b_10_SIM))
```

```python
##Ejericio 7
def poisson_TI(lamda):
    U = random()
    i = 0; p = exp(-lamda)
    F = p
    while U >= F:
        i += 1
        p *= lamda / i
        F = F + p
    return i

def poisson_mejorado(lamda):
    p = exp(-lamda); F = p
    for j in range(1, int(lamda) + 1):
        p *= lamda / j
        F += p
    U = random()
    if U >= F:
        j = int(lamda) + 1
        while U >= F:
            p *= lamda / j; F += p
            j += 1
        return j - 1
    else:
        j = int(lamda)
        while U < F:
            F -= p; p *= j/lamda
            j -= 1
        return j+1
        
Lambda      = 0.7
N = 1000

# Tiempo de la poisson con Transformada inversa sin optimizar
time_TI = time.time()
ti = 0
for _ in range(N):
    Y = poisson_TI(Lambda)
    if (Y in [0, 1, 2]):
        ti += 1

resultado_TI = np.round(1 - (ti / N), 3)
time_TI = np.round((time.time() - time_TI) * 10**3, 4)

# Tiempo de la poisson con Transformada inversa optimizada
time_TI_m = time.time()
ti_m = 0
for _ in range(N):
    Y = poisson_mejorado(Lambda)
    if (Y in [0, 1, 2]):
        ti_m += 1
resultado_TI_m = np.round(1 - (ti_m / N), 3)
time_TI_m = np.round((time.time() - time_TI_m) * 10**3, 4)

print("""
Estimacion de P(Y > 2):
Usando transformada inversa sin optimizar: {} en {} ms
Usando transformada inversa optimizada   : {} en {} ms
""".format(resultado_TI, time_TI, resultado_TI_m, time_TI_m))
```
